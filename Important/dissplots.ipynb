```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Load dataset
file_path = "FINALDATA/WTS_ES_CUBIC.csv"
df = pd.read_csv(file_path)

# Drop filename column if present
df = df.drop(columns=["filename"], errors="ignore")

# Encode 'track_type' numerically
if "track_type" in df.columns:
    encoder = LabelEncoder()
    df["track_type"] = encoder.fit_transform(df["track_type"])

# Define logical feature order (Including track_type at the top)
ordered_features = [
    "track_type",  # Include track_type first to see correlations
    "mean_log_velocity", "max_log_velocity", "min_log_velocity", "stddev_log_velocity",  # Velocity
    "mean_acceleration", "max_acceleration", "min_acceleration", "stddev_acceleration",  # Acceleration
    "mean_curvature", "max_curvature", "stddev_curvature",  # Curvature
    "reorientations_per_sec", "tortuosity", "path_length",  # Swimming patterns
    "gradient", "intercept"  # General trends
]

# Rename features for clarity
rename_dict = {
    "track_type": "Track Type",
    "gradient": "Gradient",
    "intercept": "Y-Intercept",
    "mean_log_velocity": "Mean Log Velocity",
    "stddev_log_velocity": "Std Dev Log Velocity",
    "max_log_velocity": "Max Log Velocity",
    "min_log_velocity": "Min Log Velocity",
    "mean_acceleration": "Mean Acceleration",
    "stddev_acceleration": "Std Dev Acceleration",
    "max_acceleration": "Max Acceleration",
    "min_acceleration": "Min Acceleration",
    "reorientations_per_sec": "Reorientation Rate",
    "mean_curvature": "Mean Curvature",
    "stddev_curvature": "Std Dev Curvature",
    "max_curvature": "Max Curvature",
    "tortuosity": "Tortuosity",
    "path_length": "Path Length"
}

# Reorder & rename features
df_numeric = df[ordered_features].rename(columns=rename_dict)

# Compute correlation matrix
corr_matrix = df_numeric.corr()

# Rename index and columns of the correlation matrix
corr_matrix = corr_matrix.rename(index=rename_dict, columns=rename_dict)

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5,
    cbar_kws={"shrink": 0.8}, square=True
)

# Improve title and labels
plt.title("Feature Correlation Matrix", fontsize=14)
plt.xticks(rotation=30, ha="right", fontsize=10)
plt.yticks(fontsize=10)
plt.savefig('corr_mat_WE.pdf')

plt.show()


```
