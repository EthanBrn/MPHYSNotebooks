```python
def add_missing_values_2(file_path):
    data = pd.read_csv(file_path, delim_whitespace=True, header=None, names=['time', 'x', 'y', 'z', 'x_smooth', 'y_smooth', 'z_smooth'])
    
    x = data['x']
    y = data['y']
    z = data['z']
    x_smooth = data['x_smooth']
    y_smooth = data['y_smooth']
    z_smooth = data['z_smooth']
    time = data['time']
    
    new_data = []
    new_data.append(data.iloc[0].to_dict())

    for i in range(1, len(data)-1):
        t_next = time.iloc[i+1]
        t_current = time.iloc[i]
        dt = round((t_next - t_current), 3)

        # Calculate the number of missing points
        no_of_missing_points = int((dt / 0.005) - 1)

        new_data.append(data.iloc[i].to_dict())
        time_step = 0.005
        # Only add missing points if there are any
        if no_of_missing_points > 0:
            for j in range(1, no_of_missing_points + 1):
                # Calculate the new time value
                new_time = t_current + (j * time_step)

                # Interpolate x y z data for the missing points
                new_x_smooth = x_smooth.iloc[i] + (x_smooth.iloc[i+1] - x_smooth.iloc[i]) * (j / (no_of_missing_points + 1))
                new_y_smooth = y_smooth.iloc[i] + (y_smooth.iloc[i+1] - y_smooth.iloc[i]) * (j / (no_of_missing_points + 1))
                new_z_smooth = z_smooth.iloc[i] + (z_smooth.iloc[i+1] - z_smooth.iloc[i]) * (j / (no_of_missing_points + 1))
                new_x = x.iloc[i] + (x.iloc[i+1] - x.iloc[i]) * (j / (no_of_missing_points + 1))
                new_y = y.iloc[i] + (y.iloc[i+1] - y.iloc[i]) * (j / (no_of_missing_points + 1))
                new_z = z.iloc[i] + (z.iloc[i+1] - z.iloc[i]) * (j / (no_of_missing_points + 1))

                # Create a new row for the interpolated values
                new_row = {
                    'time': new_time,
                    'x': new_x,
                    'y': new_y,
                    'z': new_z,
                    'x_smooth': new_x_smooth,
                    'y_smooth': new_y_smooth,
                    'z_smooth': new_z_smooth
                }

                # Add the new row to the data
                new_data.append(new_row)

    # Add the last row of the original data
    new_data.append(data.iloc[-1].to_dict())

    # Create a new DataFrame with the updated data
    new_df = pd.DataFrame(new_data)
    
    return new_df
```

```python


def process_txt_in_directories(directories, save_directory):
    # Make sure the save directory exists
    os.makedirs(save_directory, exist_ok=True)
    
    # Loop over each directory
    for directory in directories:
        # List all .txt files in the directory
        for filename in os.listdir(directory):
            if filename.endswith('.txt'):
                #Assign file path 
                file_path = os.path.join(directory, filename)
                
                # Call the add_missing_values function 
                processed_data = add_missing_values_2(file_path)
                
                # Create a new filename to show it is processed 
                new_filename = filename.replace('.txt', '_processed.txt')
                
                # Save the new file in save_directory
                save_path = os.path.join(save_directory, new_filename)
                # Save as txt file 
                processed_data.to_csv(save_path, sep=' ', index=False, header=False)
                #Show that it has been saved
                print(f"Processed and saved: {save_path}")
```
